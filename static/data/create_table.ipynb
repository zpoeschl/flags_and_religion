```python
import pandas as pd
```

```python
columns = ['ID', 'Country', 'Landmass', 'Zone', 'Area in Thousands of Sq. km', 'Population in Round Millions', 'Primary Language',
          'Primary Religion', '# of Bars', '# of Stripes', '# of Colors', 'Red', 'Green', 'Blue', 'Gold /Yellow',
          'White', 'Black', 'Orange /Brown', 'Main Hue', '# of Circles', '# of Crosses', '# of Saltires', '# of Quarters',
          'Sunstars', 'Crescent', 'Triangle', 'Icon', 'Animate', 'Text', 'Color, Top Left Corner', 'Color, Bottom Right Corner',
          'Area in Sq. Miles', 'Population Density']
flags_df = pd.read_csv('flags_with_headers.csv', names=columns)
```

```python
# drop columns, irrelevant or dup info
flags_df = flags_df.drop(columns=['ID', 'Landmass', 'Zone', 'Area in Thousands of Sq. km', 'Population in Round Millions',
                                  'Primary Language', 'Zone', 'Color, Top Left Corner', 'Color, Bottom Right Corner',
                                  'Area in Sq. Miles', 'Population Density'])

# drop first row (dup headers)
flags_df = flags_df.drop([0])
```

```python
flags_df
```

```python
# update 'religion' from numbers to text
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '0')] = 'Catholic'
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '1')] = 'Other Christian'
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '2')] = 'Muslim'
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '3')] = 'Buddhist'
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '4')] = 'Hindu'
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '5')] = 'Ethnic'
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '6')] = 'Marxist'
flags_df['Primary Religion'].loc[(flags_df['Primary Religion'] == '7')] = 'Other'

# update columns 1/0 to be T/F
flags_df['Red'].loc[(flags_df['Red'] == '0')] = 'False'
flags_df['Red'].loc[(flags_df['Red'] == '1')] = 'True'

flags_df['Green'].loc[(flags_df['Green'] == '0')] = 'False'
flags_df['Green'].loc[(flags_df['Green'] == '1')] = 'True'

flags_df['Blue'].loc[(flags_df['Blue'] == '0')] = 'False'
flags_df['Blue'].loc[(flags_df['Blue'] == '1')] = 'True'

flags_df['Gold /Yellow'].loc[(flags_df['Gold /Yellow'] == '0')] = 'False'
flags_df['Gold /Yellow'].loc[(flags_df['Gold /Yellow'] == '1')] = 'True'

flags_df['White'].loc[(flags_df['White'] == '0')] = 'False'
flags_df['White'].loc[(flags_df['White'] == '1')] = 'True'

flags_df['Black'].loc[(flags_df['Black'] == '0')] = 'False'
flags_df['Black'].loc[(flags_df['Black'] == '1')] = 'True'

flags_df['Orange /Brown'].loc[(flags_df['Orange /Brown'] == '0')] = 'False'
flags_df['Orange /Brown'].loc[(flags_df['Orange /Brown'] == '1')] = 'True'

flags_df['Crescent'].loc[(flags_df['Crescent'] == '0')] = 'False'
flags_df['Crescent'].loc[(flags_df['Crescent'] == '1')] = 'True'

flags_df['Triangle'].loc[(flags_df['Triangle'] == '0')] = 'False'
flags_df['Triangle'].loc[(flags_df['Triangle'] == '1')] = 'True'

flags_df['Icon'].loc[(flags_df['Icon'] == '0')] = 'False'
flags_df['Icon'].loc[(flags_df['Icon'] == '1')] = 'True'

flags_df['Animate'].loc[(flags_df['Animate'] == '0')] = 'False'
flags_df['Animate'].loc[(flags_df['Animate'] == '1')] = 'True'

flags_df['Text'].loc[(flags_df['Text'] == '0')] = 'False'
flags_df['Text'].loc[(flags_df['Text'] == '1')] = 'True'

# reformat final two columns to be capital case
flags_df['Main Hue'] = flags_df['Main Hue'].str.capitalize()

flags_df.head(20)
```
